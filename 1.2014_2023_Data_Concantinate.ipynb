```python
# Import setup
import pandas as pd
```

```python
# Create a list containing all the files for Homelessnes in US
files = ["Resources/2014-PIT-Counts-by-State.csv",
         "Resources/2015-PIT-Counts-by-State.csv",
         "Resources/2016-PIT-Counts-by-State.csv",
         "Resources/2017-PIT-Counts-by-State.csv",
         "Resources/2018-PIT-Counts-by-State.csv",
         "Resources/2019-PIT-Counts-by-State.csv",
         "Resources/2020-PIT-Counts-by-State.csv",
         "Resources/2021-PIT-Counts-by-State.csv",
         "Resources/2022-PIT-Counts-by-State.csv",
         "Resources/2023-PIT-Counts-by-State.csv"]
```

```python
Homelessnes_df = pd.DataFrame()
```

```python
for file in files:
    data = pd.read_csv(file) 
    Homelessnes_df = pd.concat([Homelessnes_df, data], axis=0)
```

```python
Homelessnes_df.tail()
```

```python
Homelessnes_df.to_csv('Resources/Homelessness_2014_2023.csv', index=False)
```
